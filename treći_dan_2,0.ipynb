```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
iris_df= pd.DataFrame (data=iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
iris_df['species']= iris.target
iris_df.head()
```

```python
species_map={0:'setosa',1:'versicolor', 2: 'virginica'}
iris_df['species']=iris_df['species'].map(species_map)
iris_df['species']
```

```python
set_versi= iris_df[iris_df['species'].isin(['stetosa','versicolor'])]
set_versi
```

```python
set_versi=iris_df[iris_df['species'].isin(['setosa', 'versicolor'])]
set_versi
```

```python
grupa1 = set_versi[set_versi['species'] == 'setosa']['sepal length (cm)']
grupa1
```

```python
grupa2 = set_versi[set_versi['species'] == 'versicolor']['sepal length (cm)']
grupa2
```

```python
#t-test usporedi arit sred grupa i postoji li značajna razlika među njima, jesu li duljine listova čašica različite, h0: prosjek je jednak, h1: prosjecna duljina jedne grupe nije jednaka drugoj, postoji stat znač razlika među njima
```

```python
ttest=pg.ttest(grupa1,grupa2)
print('T-test rezultati (st vs versi, sepal l):\n', ttest) #:\n da odbaci u novi red
```

velika t i p<0,05 značajnooo

```python
sirina = pg.ttest(iris_df[iris_df['species']=='setosa']['sepal width (cm)'],
                iris_df[iris_df['species']=='versicolor']['sepal width (cm)'])

sirina
```

```python
#anova
anova = pg.anova(dv='sepal length (cm)', between='species', data=iris_df)
anova
```

```python
corr = pg.corr(iris_df ['sepal length (cm)'], iris_df ['petal length (cm)'])
corr
```
